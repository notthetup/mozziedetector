```python
import time
import scipy, pylab
import numpy as np
import matplotlib.pyplot as plt


from numpy.fft import fft
from scipy.io import wavfile
from IPython import display
from IPython.html.widgets import interact
```

```python
pylab.rcParams['figure.figsize'] = (20.0, 8.0)
```

```python
def stft(x, fs, framesz, hop):
    """x is the time-domain signal
    fs is the sampling frequency
    framesz is the frame size, in samples
    hop is the the time between the start of consecutive frames, in samples
 
    """
    w = scipy.hamming(framesz)
    X = scipy.array([scipy.fft(w*x[i:i+framesz]) 
                     for i in range(0, len(x)-framesz, hop)])
    return X
```

```python
fs, data = wavfile.read('data/aeg_clean_norm.wav');
frameLength = 4096;
convert_16_bit = float(2**15)
normData = data/(convert_16_bit + 1.0)
```

```python
ftdata = np.absolute(stft(normData[:,1],fs,frameLength,frameLength))/frameLength;
freqdata = np.linspace(0,fs,frameLength);
```

```python
def findFundamental(data,fs,frameLength):
    harmonicSum = [sum(data[index*start] for index in range(1,5)) for start in range(0,200)]
    maxVal = max(harmonicSum);
    maxIndex = harmonicSum.index(maxVal);
    fundamentalFreq = maxIndex*fs/float(frameLength);
    return fundamentalFreq
```

```python
def plotfft(index):
    fundamental = findFundamental(ftdata[index],fs,frameLength);
    first4freq = np.multiply(range(1,5),fundamental);
    first4Index = [numpy.where(freqdata == min(freqdata, key=lambda x:abs(x-freq)))[0][0] for freq in first4freq];
    print first4Index
    plt.plot(freqdata[0:200],20*np.log10(ftdata[index][0:200]));
    plt.plot(freqdata[first4Index], 20*np.log10(ftdata[index][first4Index]), marker='o', linestyle='--', color='r')
    pylab.xticks([i*100 for i in range(0,27)])
    plt.grid();
    
interact(plotfft, index=(0,138,1));
```
