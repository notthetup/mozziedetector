```python
import time
import scipy, pylab
import numpy as np
import matplotlib.pyplot as plt


from numpy.fft import fft
from scipy.io import wavfile
from IPython import display
from IPython.html.widgets import interact
```

```python
pylab.rcParams['figure.figsize'] = (20.0, 8.0)
```

```python
def stft(x, fs, framesz, hop):
    """x is the time-domain signal
    fs is the sampling frequency
    framesz is the frame size, in samples
    hop is the the time between the start of consecutive frames, in samples
 
    """
    w = scipy.hanning(framesz)
    X = scipy.array([scipy.fft(w*x[i:i+framesz]) 
                     for i in range(0, len(x)-framesz, hop)])
    return X
```

```python
fs, data = wavfile.read('data/aeg_clean_norm.wav');
frameLength = 4096;
convert_16_bit = float(2**15)
normData = data/(convert_16_bit + 1.0)
```

```python
stftdata = np.absolute(stft(normData[:,1],fs,frameLength,frameLength))/frameLength;
freqdata = np.linspace(0,fs,frameLength);
```

```python
def findFundamental(data,fs,frameLength):
    harmonicSum = [sum(data[index*start] for index in range(1,4)) for start in range(0,200)]
    maxVal = max(harmonicSum);
    maxIndex = harmonicSum.index(maxVal);
    fundamentalFreq = maxIndex*fs/float(frameLength);
    return fundamentalFreq,maxIndex
```

```python
def findMaxAround(ftdata,first4Index):
    maxIndex = []
    searchWidth = 10;
    for fIndex in first4Index:
        sData = ftdata[fIndex-searchWidth/2:fIndex+searchWidth/2+1]
        m = max(sData);
        adjust = [i for i, j in enumerate(sData) if j == m][0]-(searchWidth/2);
        #print adjust;
        maxIndex.append(fIndex+adjust)
    return maxIndex;
```

```python
def getInterPeakIndices(first4MaxIndex):
    interPeakIdx = [first4MaxIndex[0]/2]
    diffAdj = np.diff(first4MaxIndex)/2;
    interPeakIdx.extend([diffIdx+first4MaxIndex[idx] for idx,diffIdx in enumerate(diffAdj)])
    interPeakIdx.append(diffAdj[-1]+first4MaxIndex[-1])
    return interPeakIdx
```

```python
def plotfft(index):
    ftdata = 20*np.log10(stftdata[index]);
    fFreq,fIndex = findFundamental(ftdata,fs,frameLength);

    first4Index = np.multiply(range(1,5),fIndex);

    first4MaxIndex = findMaxAround(ftdata,first4Index);
    
    interPeakIdx = getInterPeakIndices(first4MaxIndex);
    
    plt.plot(freqdata[0:200],ftdata[0:200]);
    plt.plot(freqdata[first4MaxIndex], ftdata[first4MaxIndex], marker='o', linestyle='--', color='r')
    plt.plot(freqdata[interPeakIdx], ftdata[interPeakIdx], marker='o', linestyle='--', color='g')
    pylab.xticks([i*100 for i in range(0,27)])
    plt.grid();
    
interact(plotfft, index=(0,len(stftdata),1));
```
